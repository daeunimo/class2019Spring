```python
tokens = nltk.word_tokenize(raw) #어절 단위로 잘라 리스트 형태로 리턴
print(type(tokens))
print(len(tokens))
print(tokens[:10])
```

```python
lyrics_tokens=nltk.word_tokenize(lyrics)
print(lyrics_tokens[100:250])
print(len(lyrics_tokens))
```

```python
text = nltk.Text(tokens) #nltk 전용 변수를 만들면 더 다양한 함수 활용 가능
print(type(text))
print(text[:10])
text.collocations()
```

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
# gutengerg 폴더에서 파일 불러와 raw 변수에 대입
fdist = nltk.FreqDist(raw) #빈도수
print(fdist)
print(fdist.most_common(5)) #상위 5개 출력. 리스트안의 튜플 형태.

fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
# raw 리스트의 원소들은 알페인 것을 찾아 소문자 변환한 뒤 빈도수 출력
print(fdist.most_common(5))
[char for (char, count) in fdist.most_common(5)]b
```

```python
lyrics_fdist=nltk.FreqDist(text2)
print(lyrics_fdist.most_common(20))
lyrics_fdist= nltk.FreqDist(ch2*5 for ch2 in text2 if ch2=='사랑')
print(lyrics_fdist.most_common(5))
```
